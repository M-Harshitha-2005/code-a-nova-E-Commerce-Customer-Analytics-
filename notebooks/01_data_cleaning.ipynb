## Data Cleaning and preprossing

```python
import pandas as pd
import numpy as np
```

### loading dataset

```python
df = pd.read_csv("../data/online_retail_II.csv")
df.head()
```

### data shape

```python
print("Rows and Columns:")
print(df.shape)
```

```python
df.info()
```

### Missing values

```python
df.isnull().sum()
```

### Duplicate values

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.describe()
```

### sales column

```python
df["Sales"] = df["Quantity"] * df["Price"]
df.head()
```

### sales statics

```python
df["Sales"].describe()
```

### conversion of invoice date

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["InvoiceDate"].dtype
```

```python
df.dtypes
```

### Datacleaning

#### removing missing customerIds

```python
print("Rows Before:", df.shape)
df = df.dropna(subset=["Customer ID"])
print("Rows After:", df.shape)
```

```python
df = df.dropna(subset=["Description"])
df.shape
```

#### remove duplicate records

```python
print("Before:", df.shape)
df = df.drop_duplicates()
print("After:", df.shape)
```

#### removing cancelled orders

```python
df = df[~df["Invoice"].astype(str).str.startswith("C")]
df.shape
```

#### removing -ve quantites

```python
df = df[df["Quantity"] > 0]
df.shape
```

#### removing -ve prices

```python
df = df[df["Price"] > 0]
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Sales"].describe()
```

### data features

```python
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["Day"] = df["InvoiceDate"].dt.day
df.head()
```

```python
# verifying new columns
df[["InvoiceDate", "Year", "Month", "Day"]].head()
```

### data set overview

```python
print("Final Shape:", df.shape)
```

saving the clean dataset

```python
df.to_csv("../data/cleaned_data.csv", index=False)
print("Cleaned dataset saved successfully.")
```
